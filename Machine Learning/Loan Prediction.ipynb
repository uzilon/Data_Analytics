## **Project Implementation**

## Data Cleaning and Filling Missing Values

### Missing Values

```python
# To find how many missing values are in the dataset

df.isnull()
```

```python
# find total number of missing values in each column

total = df.isnull().sum()
total
```

```python
# Access count of gender
df['Gender'].value_counts()
```

```python
# Fill Missing values with Male 

df.Gender = df.Gender.fillna('Male')
```

```python
# Access count of Married
df['Married'].value_counts()
```

```python
# Fill Missing values with Yes 

df.Married = df.Married.fillna('Yes')
```

```python
# Access count of Dependents
df['Dependents'].value_counts()
```

```python
# Fill Missing values with 0 

df.Dependents = df.Dependents.fillna(0)
```

```python
# Access count of Self-employed
df['Self_Employed'].value_counts()
```

```python
# Fill Missing values with No 

df.Self_Employed = df.Self_Employed.fillna('No')
```

```python
# Access count of Loan Amount Term
df['Loan_Amount_Term'].value_counts()
```

```python
# Fill Missing values with 360.0

df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(360.0)
```

```python
# Access count of Credit History
df['Credit_History'].value_counts()
```

```python
# Fill Missing values with 1.0

df.Credit_History = df.Credit_History.fillna(1.0)
```

```python
# For Loan amount (numeric variable), we fill with the mean value
mean = df.LoanAmount.mean()
df.LoanAmount = df.LoanAmount.fillna(mean)
```

```python
# rerun code after filling missing values
total = df.isnull().sum()
total
```

## Converting Categorical Variable into Numeric One - Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
le = LabelEncoder()

for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
df["Dependents"] = df["Dependents"].replace('3+', 3)
```

```python
df.head(30)
```

## Feature Selection



```python
x = df.iloc[:, 1:12]
y = df.iloc[:, 12] 
```

```python
x.head(30)
```

```python
y.head()
```

## Spliting Data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state = 0)

x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Building Decison Tree Models

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()

model.fit(x_train, y_train)

y_predictions = model.predict(x_test)
```

## Evaluating Model

```python
from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_test, y_predictions))
```

##### Accuracy is 70.73%

```python
#Using Confusion Metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_predictions)
cnf_matrix
```

##### 22 & 65 are correct predictions
##### 11 & 25 are incorrect predictions

##### Out of 123, 87 are correct and 36 are incorrect predictions

## Logistic Regression Model (To Get a Higher Accuracy)

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred
```

## Evaluating Model

```python
from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

##### Accuracy is 83.73%

```python
#Using Confusion Metrics
cnf_matrix2 = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix2
```

##### 15 & 88 are correct predictions
##### 2 & 18 are incorrect predictions

##### Out of 123, 103 are correct and 20 are incorrect predictions

## Visualizing Confusion Matrix Using Heat Map

```python
%matplotlib inline
import seaborn as sn
sn.heatmap(cnf_matrix2, annot = True)
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show
```
